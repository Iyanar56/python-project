```python
sns.countplot(data = df, x ='Sex')
plt.title("Distrubition By Genter")
plt.show()
```

```python
sns.histplot(data = df, x = "Age", bins = 10,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data = df, x = "Height", bins = 20,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data = df, x = "Weight", bins = 20,kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data = df, x = "Medal")
plt.title("Medals Distribution")
plt.show()
```

```python
sns.countplot(data = df, x = "Year", hue = "Medal")
plt.title("Medal Distribution Over The Years")
plt.xticks(rotation = 50)
plt.show()
```

```python
year_avg_age = df.groupby("Year")["Age"].mean()
print(year_avg_age)
```

```python
sport_median_height = df.groupby("Sport")["Height"].median()
print(sport_median_height.max())
```

```python
sport_median_height[sport_median_height == 190.0]
```

```python
country_gender_count = df.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
country_gold_medals = df[df["Medal"] == "Gold"].groupby("NOC")["Medal"].count()
print(country_gold_medals)
```

```python
country_gold_medals = df[df["Medal"] == "Gold"].groupby("NOC")["Medal"].count()
print(country_gold_medals.max())
```

```python
country_gold_medals[country_gold_medals == 2638]
```

```python
sport_gendar_avg_weight = df.groupby(["Sport","Sex"])["Weight"].mean()
print(sport_gendar_avg_weight["Wrestling"])
```

```python
year_participant_count = df.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind = "bar")
plt.title("Number Of Participants Over The Year")
plt.xlabel("Year")
plt.ylabel("Number Of Participants")
plt.show()
```

```python
country_avg_age = df.groupby("NOC")["Age"].mean().sort_values(ascending = False)
country_avg_age.head(10).plot(kind = "bar")
plt.title("Top 10 Countries With Highest Average Age Of Participants")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.show()
```

```python
country_avg_age = df.groupby("NOC")["Age"].mean().sort_values(ascending = False)
country_avg_age.tail(15).plot(kind = "bar")
plt.title("Top 15 Countries With Lowers Average Age Of Participants")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.show()
```

```python
sns.boxplot(data = df, x = "Season",y = "Age")
plt.title("Distribution of Ages By Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
most_medals_county = df["NOC"].value_counts().idxmax()
print("Most Medals-Winning Country: ", most_medals_county)
```

```python
tallest_athlete = df[df["Height"] == df["Height"].max()]
print("Tallest Athlete:")
print(tallest_athlete[["ID","Name","Height","Sport"]])
```

```python
heaviest_athlete = df[df["Weight"] == df["Weight"].min()]
print("Heaviest Athlete:")
print(tallest_athlete[["ID","Name","Weight","Sport"]])
```

```python
sns.scatterplot(data = df, x = "Height",y = "Weight",hue = "Medal",palette = "Set1")
plt.title("Athlete Height vs Weight by Medal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
medals_by_country_year = df.pivot_table(index = "NOC",columns = "Year",values = "Medal",aggfunc="count")
sns.heatmap(medals_by_country_year,linewidths = 0.5)
plt.title("Medals Counts by Country And year")
plt.xlabel("Year")
plt.ylabel("Country")
plt.show()
```
